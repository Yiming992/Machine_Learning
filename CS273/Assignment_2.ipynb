# Problem 1

```python
#(a)

X = curve[:,0]
X = X[:,np.newaxis]
Y=curve[:,1] 
Xtr,Xte,Ytr,Yte = ml.splitData(X,Y,0.75) # split data set 75/25
Ytr=np.reshape(Ytr,(60,1))
Yte=np.reshape(Yte,(20,1))
```

```python
#(b)

lr = ml.linear.linearRegress( Xtr, Ytr ); # create and train model
xs = np.linspace(0,10,200);# densely sample possible x-values
xs = xs[:,np.newaxis]# force "xs" to be an Mx1 matrix 
ys = lr.predict( xs );# make predictions at xs
```

```python
%matplotlib inline
plt.scatter(Xtr,Ytr)
Ypretr=lr.predict(Xtr)# Predicted values based on training data
plt.plot(xs,ys,c='red') # Red line represents the estimated prediction function
plt.xlabel('X')
plt.ylabel('Y')
```

```python
print('Coefficients: {}'.format(lr.theta))
```

```python
# Mean squared error on training set
np.sum((Ypretr-Ytr)**2)/Xtr.shape[0]
```

```python
# Mean squared error on test set
Ypret=lr.predict(Xte)
np.sum((Ypret-Yte)**2)/Xte.shape[0]
```

```python
# (c)
DEGREE=[1,3,5,7,10,18]
MSE_TE=[]
MSE_TR=[]



# Degree d=1

XtrP = ml.transforms.fpoly(Xtr, 1, bias=False)
# Rescasle the data matrix so that the features have similar ranges / variance
XtrP,params = ml.transforms.rescale(XtrP)
# Then we can train the model on the scaled feature matrix:
lr = ml.linear.linearRegress( XtrP, Ytr ); # create and train model
# Now, apply the same polynomial expansion & scaling transformation to Xtest and xs
XteP,_= ml.transforms.rescale( ml.transforms.fpoly(Xte,1,bias=False), params);
xsP,_= ml.transforms.rescale( ml.transforms.fpoly(xs,1,bias=False), params)
ysP= lr.predict(xsP)
# Plot for the estimated regression function
plt.scatter(XtrP,Ytr)
plt.plot(xsP,ysP,c='red')
ax=plt.axis()
plt.xlabel('X')
plt.ylabel('Y')

# MSE


Ypretr=lr.predict(XtrP)
# Mean squared error on training set
msetr_1=np.sum((Ypretr-Ytr)**2)/XtrP.shape[0]
# Mean squared error on test set
Ypret=lr.predict(XteP)
msete_1=np.sum((Ypret-Yte)**2)/XteP.shape[0]

MSE_TR.append(msetr_1)
MSE_TE.append(msete_1)
```

# Problem 2

```python
# Conduct 5-fold cross validation

DEGREE=[1,3,5,7,10,18]
nFolds = 5;
CV=np.empty((6,5))

    
for iFold in range(nFolds):
    Xti,Xvi,Yti,Yvi = ml.crossValidate(Xtr,Ytr,nFolds,iFold)# take ith data block as learner
    i=0
    for degree in DEGREE:
        Yti=np.reshape(Yti,(Yti.shape[0],1))
        Yvi=np.reshape(Yvi,(Yvi.shape[0],1))
        XtiP = ml.transforms.fpoly(Xti, degree, bias=False)
        XtiP,params = ml.transforms.rescale(XtiP)
        XviP,_=ml.transforms.rescale(ml.transforms.fpoly(Xvi, degree, bias=False),params)
        lr = ml.linear.linearRegress( XtiP, Yti )
        YviP=lr.predict(XviP)
        CV[i,iFold]=np.sum((YviP-Yvi)**2/XviP.shape[0])
        i=i+1
CV=np.mean(CV,axis=1)# Cross-validation errors for various degrees
    
    
```

```python

# Plot of cross validation error versus degree of polynomial
plt.semilogy(DEGREE,CV,c='red')
plt.xlabel('Degree')
plt.ylabel('CV-MSE')
```

```python
# Cross validation errors
print('MSE Cross Validation:{}'.format(CV))
```

From above results, minimum cross validation error is obtained when degree d=5.

MSEs estimated from cross-validation are generally close to the MSEs estimated from previous test data, except for when d=18, in which case MSE estimated from cross-validation is significantly larger than that estimated through test data.
