```python
#(a)
iris = np.genfromtxt("/Users/yimingliu/Downloads/HW1-code/data/iris.txt",delimiter=None) # load the data 
Y = iris[:,-1]
X = iris[:,0:2]# Using only first two features
import mltools as ml
np.random.seed(0)
X,Y = ml.shuffleData(X,Y); # shuffle data randomly
Xtr,Xte,Ytr,Yte = ml.splitData(X,Y, 0.75); # split data into 75/25 train/test
```

```python
# K=1
knn = ml.knn.knnClassify() # create the object and train it
knn.train(Xtr, Ytr, 1) # where K is an integer, e.g. 1 for nearest neighbor prediction 
YteHat = knn.predict(Xte) # get estimates of y for each data point in Xte
ml.plotClassify2D( knn, Xtr, Ytr ); # make 2D classification plot with data (Xtr,Ytr)
```

```python
# K=5
knn = ml.knn.knnClassify() # create the object and train it
knn.train(Xtr, Ytr, 5) # where K is an integer, e.g. 1 for nearest neighbor prediction 
YteHat = knn.predict(Xte) # get estimates of y for each data point in Xte
ml.plotClassify2D( knn, Xtr, Ytr ); # make 2D classification plot with data (Xtr,Ytr)
```

```python
# K=10
knn = ml.knn.knnClassify() # create the object and train it
knn.train(Xtr, Ytr, 10) # where K is an integer, e.g. 1 for nearest neighbor prediction 
YteHat = knn.predict(Xte) # get estimates of y for each data point in Xte
ml.plotClassify2D( knn, Xtr, Ytr ); # make 2D classification plot with data (Xtr,Ytr)
```

```python
# K=50
knn = ml.knn.knnClassify() # create the object and train it
knn.train(Xtr, Ytr, 50) # where K is an integer, e.g. 1 for nearest neighbor prediction 
YteHat = knn.predict(Xte) # get estimates of y for each data point in Xte
ml.plotClassify2D( knn, Xtr, Ytr ); # make 2D classification plot with data (Xtr,Ytr)
```

```python
#(b)
K=[1,2,5,10,50,100,200]
errTrain=[]
errte=[]
for k in K:
    learner=ml.knn.knnClassify(Xtr,Ytr,k)
    Yhattr=learner.predict(Xtr)
    Yhatte=learner.predict(Xte)
    errTrain.append(1-(np.sum(Ytr==Yhattr)/Ytr.shape[0]))
    errte.append(1-(np.sum(Yte==Yhatte)/Yte.shape[0]))
```

```python
plt.semilogx(K,errTrain,c='red')
plt.semilogx(K,errte,c='green')
plt.legend(['Training error', 'Test error'], loc='upper left')
```

From above plots, we can see test error, which represents the general performance of the classifier, reaches the lowest point when k=50. Thus I would recommend 50 as the most appropriate k in our settings.

(b)

For X=(0 0 0 0 0):

\begin{align}
P(Y=1|X) & \propto P(X|Y=1)P(Y=1) \\\\[5pt]
 &\propto P(X_{1}=0|Y=1)P(X_{2}=0|Y=1)P(X_{3}=0|Y=1)P(X_{4}=0|Y=1)P(X_{5}=0|Y=1)P(Y=1)\\\\[5pt]
 & \propto 1/4 \cdot 1\cdot 1/4 \cdot 1/2 \cdot 3/4 \cdot 2/5\\\\
 & \propto 3/320
\end{align}

\begin{align}
P(Y=-1|X) & \propto P(X|Y=-1)P(Y=-1) \\\\[5pt]
 &\propto P(X_{1}=0|Y=-1)P(X_{2}=0|Y=-1)P(X_{3}=0|Y=-1)P(X_{4}=0|Y=-1)P(X_{5}=0|Y=-1)P(Y=-1)\\\\[5pt]
 & \propto 1/2 \cdot 1/6\cdot 1/3 \cdot 1/6 \cdot 2/3 \cdot 3/5\\\\
 & \propto 1/540
\end{align}

Since 3/320 > 1/540
Class 'Read' should be predicted for X=(0 0 0 0 0 )

For X=(1 1 0 1 0):
\begin{align}
P(Y=1|X) & \propto P(X|Y=1)P(Y=1) \\\\[5pt]
 &\propto P(X_{1}=1|Y=1)P(X_{2}=1|Y=1)P(X_{3}=0|Y=1)P(X_{4}=1|Y=1)P(X_{5}=0|Y=1)P(Y=1)\\\\[5pt]
 & \propto 3/4 \cdot 0\cdot 1/4 \cdot 1/2 \cdot 3/4 \cdot 2/5\\\\
 & \propto 0
\end{align}

\begin{align}
P(Y=-1|X) & \propto P(X|Y=-1)P(Y=-1) \\\\[5pt]
 &\propto P(X_{1}=1|Y=-1)P(X_{2}=1|Y=-1)P(X_{3}=0|Y=-1)P(X_{4}=1|Y=-1)P(X_{5}=0|Y=-1)P(Y=-1)\\\\[5pt]
